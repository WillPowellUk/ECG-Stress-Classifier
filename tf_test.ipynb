```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
# from tensorflow.keras.metrics import binary_focal_crossentropy

class Dataset:
    def __init__(self):
        # Generate random data for training and testing
        self.X_train = np.random.rand(1000, 800)
        self.y_train = np.random.randint(0, 3, size=(1000,))
        self.X_test = np.random.rand(200, 800)
        self.y_test = np.random.randint(0, 3, size=(200,))
        
    def cnn(self):
        # Instantiate the Dataset class
        self.dataset = Dataset()

        verbose, epochs, batch_size = 0, 10, 32

        model = keras.Sequential()
        model.add(layers.Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(800, 1)))
        model.add(layers.MaxPooling1D(pool_size=2))
        model.add(layers.Flatten())
        model.add(layers.Dense(units=3, activation='softmax'))

        # compile model
        model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

        # fit model
        model.fit(self.dataset.X_train.reshape(-1, 800, 1), self.dataset.y_train, epochs=epochs, batch_size=batch_size, validation_data=(self.dataset.X_test.reshape(-1, 800, 1), self.dataset.y_test))


data = Dataset()
data.cnn()
```
